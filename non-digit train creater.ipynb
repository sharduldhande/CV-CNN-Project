```python
import pickle
import random

import cv2
from MSER import aspect_filter
from MSER import mser_detection
from MSER import nms_filter
```

```python
bboxes_path = 'Data/train/bboxes.pkl'
names_path = 'Data/train/names.pkl'

with open(bboxes_path, 'rb') as file:
    original_bboxes = pickle.load(file)
with open(names_path, 'rb') as file1:
    original_names = pickle.load(file1)
```

```python
def is_digit(givenbox, actualboxes):
    x1, y1, w1, h1, = givenbox

    for box in actualboxes:
        x2 = box['left']
        y2 = box['top']
        w2 = box['width']
        h2 = box['height']

        if x1 < x2 + w2 and x1 + w1 > x2 and y1 < y2 + h2 and y1 + h1 > y2:
            return True

    return False
```

```python
nondigiboxes=[]
nondiginames=[]
nondigiindices=[]
```

```python
from MSER import mser_pipeline
import os
folder_path = 'Data/train'
image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.png'))]



used = set()
count = 0

while len(nondigiboxes) < 5000:
    randomimage = random.choice(image_files)
    if randomimage in used:
        continue
    used.add(randomimage)
    image_path = os.path.join(folder_path, randomimage)
    # print(image_path)
    img = cv2.imread(image_path)
    bboxes = mser_pipeline(img)
    if bboxes is None:
        continue
    abbox = random.choice(bboxes)

    index = int(''.join(char for char in randomimage if char.isdigit())) -1

    isDigit = is_digit(abbox, original_bboxes[index])
    x1, y1, w1, h1, = abbox
    if not isDigit:
        bbox = {}
        bbox['left'] = x1
        bbox['top'] = y1
        bbox['width'] = w1
        bbox['height'] = h1
        bbox['label'] = 10
        nondigiboxes.append([bbox])
        aname = str(index+1) + ".png"
        nondiginames.append(aname)
        nondigiindices.append(index)

        count+=1
        # print(count)
        if count%500 == 0:
            print(count)
```
