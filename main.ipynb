```python
import cv2
from MSER import aspect_filter, resolutionizer
from MSER import mser_detection
from MSER import nms_filter
```

```python
testimagepath = 'Data/test5.jpg'
test_img = cv2.imread(testimagepath)
```

```python
image = resolutionizer(test_img)
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

bboxes = mser_detection(image)
print("before aspect filter:", len(bboxes))
bboxes = aspect_filter(bboxes)
print("after aspect filter:", len(bboxes))
bboxes = nms_filter(gray_img, bboxes)
print("after nms filter:", len(bboxes))
```

```python
from MSER import mser_pipeline

bboxes2 = mser_pipeline(image)
print("after pipeline filter:", len(bboxes2))
```

```python
image1 = image.copy()
for (x,y,w,h) in bboxes:
    cv2.rectangle(image1, (x,y), (x+w, y+h), (0,0,255), 1)

cv2.imshow("bboxes",image1)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```
