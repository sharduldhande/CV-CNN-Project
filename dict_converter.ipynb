```python
import mat73
import pickle

file_path = 'Data/test/digitStruct.mat'
data = mat73.loadmat(file_path)

print("loading done")
```

```python
print(type(data))
```

```python
bboxes = []
names = []

digit_struct = data['digitStruct']


for i in range(len(digit_struct['bbox'])):
    if i %1000 == 0: print(i)
    bbox = digit_struct['bbox'][i]
    name = digit_struct['name'][i]

    names.append(name)

    image_bboxes = []

    if isinstance(bbox['height'], list):
        num_digits = len(bbox['height'])
    else:
        num_digits = 1

    for j in range(num_digits):
        if num_digits == 1 and not isinstance(bbox['height'], list):
            digit_dict = {
                'height': float(bbox['height']),
                'width': float(bbox['width']),
                'top': float(bbox['top']),
                'left': float(bbox['left']),
                'label': int(bbox['label'])
            }
        else:
            digit_dict = {
                'height': float(bbox['height'][j]),
                'width': float(bbox['width'][j]),
                'top': float(bbox['top'][j]),
                'left': float(bbox['left'][j]),
                'label': int(bbox['label'][j])
            }

        image_bboxes.append(digit_dict)

    bboxes.append(image_bboxes)
```

```python
print(bboxes[11])
names[11]
```

```python
with open('Data/test/bboxes.pkl', 'wb') as f:
    pickle.dump(bboxes, f)

with open('Data/test/names.pkl', 'wb') as f:
    pickle.dump(names, f)
```
